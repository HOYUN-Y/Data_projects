#### 단속일시정보 컬럼 전처리
object ==> datetime64

#### 추가 컬럼
- 단속일시(요일) 
- 단속일시(평일/주말) 
- 단속일시(월)
- 단속일시(계절)
- 단속일시(시간)
- 단속일시(오전/오후)

```python
illigal_parking = illigal_parking.astype({'단속일시정보':'datetime64[ns]'})
illigal_parking
```

```python
illigal_parking['단속일시_월'] = illigal_parking['단속일시정보'].dt.month
illigal_parking['단속일시_연'] = illigal_parking['단속일시정보'].dt.year
illigal_parking['단속일시_일'] = illigal_parking['단속일시정보'].dt.day
illigal_parking['단속일시_요일'] = illigal_parking['단속일시정보'].dt.day_name()
```

```python
illigal_parking['단속일시_시'] = illigal_parking['단속일시정보'].dt.hour
```

```python
illigal_parking['단속일시_isweekday'] = illigal_parking['단속일시_요일'].apply(lambda x: '주말' if x >= 5 else '평일')
```

```python
def daytime(x):
    if x>=6 and x<=12:
        return 'morning'
    if x>12 and x<18:
        return 'afternoon'
    else:
        return 'night'
```

```python
illigal_parking['단속일시_isdaytime'] = illigal_parking['단속일시_시'].apply(daytime)
```

```python
illigal_parking['단속일시_season'] = illigal_parking['단속일시_월'].apply(lambda x: 
                                                                 'spring' if x in [3,4,5] else
                                                                 'summer' if x in [6,7,8] else
                                                                 'fall' if x in [9,10,11] else
                                                                 'winter'
                                                                 )
```

### 수원시 불법주정차 카메라 위치 정보
https://www.suwon.go.kr/web/deptHome/dep_traffic/traffic02/parking/pages/ExtParkJuchaInfoBub.do#none

```python
import requests
import pandas as pd
```

```python
cctv_loc_tm = pd.DataFrame(columns=['equpSn', 'admiCd', 'sectCd', 'sectCdNm', 'regltType', 'regltTypeNm',
       'cctvSn', 'instCmpy', 'instDong', 'instPont', 'instDt', 'gpsCrdntX',
       'gpsCrdntY', 'flStreNm', 'flExtsn', 'flStrePath', 'flSize', 'frstFlNm',
       'statCd'],)
```

```python
cctv_loc_tm
```

```python
gu = ['g','y','j','p'] #권선구, 영통구, 장안구, 팔달구
gu_name_map = {
    'g': '권선구', 'y': '영통구', 'j': '장안구', 'p': '팔달구'
}
all_data = []

for loc in gu:
    url = f'https://www.suwon.go.kr/web/deptHome/dep_traffic/traffic02/parking/apis/getFixCctvList.do?parId={loc}'
    params = {
        'gubun' : 'g',
        'gu':loc
    }

    headers = {
        'Content-Type' : 'application/x-www-form-urlencoded; charset=UTF-8',
        'User-Agent' : "Mozilla/5.0"
    }

    response = requests.post(url, data=params, headers = headers)
    print(f'{loc} response >>> {response}')
    
    if response.status_code == 200:
        data = response.json()

        data_df_temp = pd.DataFrame(data)
        data_df_temp['gu'] = gu_name_map[loc]
        all_data.append(data_df_temp)
    
    else:
        print(f'{loc} no response')

cctv_loc_tm = pd.concat(all_data, ignore_index=True)
    
```

```python

for i, val in enumerate(cctv_loc_tm['gpsCrdntX']):
    try:
        float(val)
    except:
        print(f"[{i}] 변환 실패: {val}")
```

```python
cctv_loc_tm['gpsCrdntX'] = cctv_loc_tm['gpsCrdntX'].astype(float)
cctv_loc_tm['gpsCrdntY'] = cctv_loc_tm['gpsCrdntY'].astype(float)
```

- gpsCrdntX : 경도(longitude)
- gpsCrdntY : 위도(Latitude)

```python
import folium
import pandas as pd

cnt = 0
m = folium.Map(location=[cctv_loc_tm['gpsCrdntX'].mean(), cctv_loc_tm['gpsCrdntY'].mean() ], zoom_start=13)

for idx, row in cctv_loc_tm.iterrows():
    lat = row['gpsCrdntX']
    lng = row['gpsCrdntY']
    name = row['flStrePath'] if 'flStrePath' in cctv_loc_tm.columns else 'CCTV'
    cnt = cnt + 1
    print(cnt , lat,lng)
    folium.Marker(
        location = [lat, lng],
        popup=name,
        icon = folium.Icon(color='blue', icon='camera', prefix='fa')
    ).add_to(m)


m.save('cctv_map.html')
```
